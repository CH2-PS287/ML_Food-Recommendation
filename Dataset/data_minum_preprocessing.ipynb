```python
# Create function to upload & download file in google colab, saved at content folder
import os
from google.colab import files

def uploadFile():
    uploaded = files.upload()
    for filename, content in uploaded.items():
        with open('/content/' + filename, 'wb') as f:
            f.write(content)

    os.listdir('/content')

def download_file(file_path):
    files.download(file_path)
```

```python
# First, upload recipe data
uploadFile()
```

```python
# Import pandas
import pandas as pd

# Read JSON data into DataFrame
recipes = pd.read_json('/content/minum.json')

# Show food name & ingredient data
for index, row in recipes.iterrows():
    # Show food name
    print(row['name'])
    for ingredient in row['food_ingredients_info']:
        # Show ingredient name
        print(f"Bahan: {ingredient.get('ingredients_desc', 'N/A')}, Kuantitas: {ingredient.get('quantity', 'N/A')}, Ukuran: {ingredient.get('measurement', 'N/A')}")
```

```python
# Proses data (move ingredient data of 'ingredients_group' to 'food_ingredients_info')
for index, row in recipes.iterrows():
    updated_ingredients = []
    for ingredient in row['food_ingredients_info']:
        # If there's data ingredients_group, then put the data to updated_ingredients list
        if 'ingredients_group' in ingredient:
            updated_ingredients.extend(ingredient['ingredients_group'])
        else:
            updated_ingredients.append(ingredient)
    # Update food_ingredients_info with updated_ingredients
    recipes.at[index, 'food_ingredients_info'] = updated_ingredients

# Remove spaces from 'ingredients_desc'
recipes['food_ingredients_info'] = recipes['food_ingredients_info'].apply(lambda x: [{k: v.strip() if k == 'ingredients_desc' else v for k, v in ingredient.items()} for ingredient in x])

# Show ingredients data after processing
for index, row in recipes.iterrows():
    print(row['name'])
    for ingredient in row['food_ingredients_info']:
        print(f"Nama bahan: {ingredient.get('ingredients_desc', 'N/A')}, Kuantitas: {ingredient.get('quantity', 'N/A')}, Ukuran: {ingredient.get('measurement', 'N/A')}")

# Save DataFrame to JSON file after processing
# data.to_json('/content/processed_data.json', orient='records', lines=False)
```

```python
# Create a list to store all food_ingredients_info
all_ingredients = []

for index, row in recipes.iterrows():
    all_ingredients.extend(row["food_ingredients_info"])

df_all_ingredients = pd.DataFrame(all_ingredients)

# Show the result
df_all_ingredients
```

```python
# Looping all recipes
for index, row in recipes.iterrows():
    # Looping food_ingredients_info
    for ingredient in row['food_ingredients_info']:
        # Looping all ingredient with nutrition
        for i, nutrition in all_ingredients_with_nutrition.iterrows():
            # if ingredients_desc, ingredients_id, quantity from food_ingredients_info is equal to
            # ingredients_desc, ingredients_id, quantity from nutrition, then add column nutrition to food_ingredients_info
            if ingredient['ingredients_desc'] == nutrition['ingredients_desc'] and ingredient['ingredients_id'] == nutrition['ingredients_id'] and ingredient['quantity'] == nutrition['quantity']:
               ingredient['air'] = nutrition['air']
               ingredient['energi'] = nutrition['energi']
               ingredient['protein'] = nutrition['protein']
               ingredient['lemak'] = nutrition['lemak']
               break
```

```python
# Accumulation nutrition from ingredient to food
for index, row in recipes.iterrows():
    # Initial variable to store accumulation of nutrition
    total_air = 0
    total_energi = 0
    total_protein = 0
    total_lemak = 0
    for ingredient in row['food_ingredients_info']:
        total_air += ingredient.get('air')
        total_energi += ingredient.get('energi')
        total_protein += ingredient.get('protein')
        total_lemak += ingredient.get('lemak')
    # Add new columns of nutrition to recipes
    recipes.at[index, 'air'] = total_air
    recipes.at[index, 'energi'] = total_energi
    recipes.at[index, 'protein'] = total_protein
    recipes.at[index, 'lemak'] = total_lemak
```
