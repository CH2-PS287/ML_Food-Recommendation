```python
# Create function to upload & download file in google colab, saved at content folder
import os
from google.colab import files

def uploadFile():
    uploaded = files.upload()
    for filename, content in uploaded.items():
        with open('/content/' + filename, 'wb') as f:
            f.write(content)

    os.listdir('/content')

def download_file(file_path):
    files.download(file_path)
```

```python
# Import pandas
import pandas as pd

# Read JSON data into DataFrame
recipes = pd.read_json('/content/snack, lauk.json')

# Show food name & ingredient data
for index, row in recipes.iterrows():
    # Show food name
    print(row['name'])
    for ingredient in row['food_ingredients_info']:
        # Show ingredient name
        print(f"Bahan: {ingredient.get('ingredients_desc', 'N/A')}, ID: {ingredient.get('ingredients_id', 'N/A')}, Kuantitas: {ingredient.get('quantity', 'N/A')}, Ukuran: {ingredient.get('measurement', 'N/A')}")
```

```python
# Proses data (move ingredient data of 'ingredients_group' to 'food_ingredients_info')
for index, row in recipes.iterrows():
    updated_ingredients = []
    for ingredient in row['food_ingredients_info']:
        # If there's data ingredients_group, then put the data to updated_ingredients list
        if 'ingredients_group' in ingredient:
            updated_ingredients.extend(ingredient['ingredients_group'])
        else:
            updated_ingredients.append(ingredient)
    # Update food_ingredients_info with updated_ingredients
    recipes.at[index, 'food_ingredients_info'] = updated_ingredients

# Remove spaces from 'ingredients_desc'
recipes['food_ingredients_info'] = recipes['food_ingredients_info'].apply(lambda x: [{k: v.strip() if k == 'ingredients_desc' else v for k, v in ingredient.items()} for ingredient in x])

# Show ingredients data after processing
for index, row in recipes.iterrows():
    print(row['name'])
    for ingredient in row['food_ingredients_info']:
        print(f"Nama bahan: {ingredient.get('ingredients_desc', 'N/A')}, ID: {ingredient.get('ingredients_id', 'N/A')}, Kuantitas: {ingredient.get('quantity', 'N/A')}, Ukuran: {ingredient.get('measurement', 'N/A')}")

# Save DataFrame to JSON file after processing
# data.to_json('/content/processed_data.json', orient='records', lines=False)
```

```python
def convert_to_float(value):
    try:
        cleaned_value = value.replace(',', '.')
        return float(cleaned_value) if cleaned_value != '-' else 0.0
    except ValueError:
        return 0.0
```

```python
# Accumulation nutrition from ingredient to food
for index, row in recipes.iterrows():
    # Initial variable to store accumulation of nutrition
    total_air = 0
    total_energi = 0
    total_protein = 0
    total_lemak = 0
    total_abu = 0
    total_karbohidrat = 0
    total_serat_total = 0
    total_gula_total = 0
    total_kalsium = 0
    total_besi = 0
    total_magnesium = 0
    total_fosfor = 0
    total_kalium = 0
    total_natrium = 0
    total_seng = 0
    total_temBaga = 0
    total_mangan = 0
    total_selenium = 0
    total_vitamin_c = 0
    total_tiamina = 0
    total_riboflavin = 0
    total_niasin = 0
    total_pantotenat = 0
    total_vitamin_b6 = 0
    total_folat = 0
    total_kolina = 0
    total_vitamin_b12 = 0
    total_vitamin_a_iu = 0
    total_vitamin_a_rae = 0
    total_retinol = 0
    total_a_karoten = 0
    total_b_karoten = 0
    total_b_kriptosantin = 0
    total_likopen = 0
    total_zeaksantin_lutein = 0
    total_vitamin_e = 0
    total_vitamin_d = 0
    total_vitamin_d_iu = 0
    total_vitamin_k = 0
    total_lemak_jenuh = 0
    total_lemak_tunggal = 0
    total_lemak_ganda = 0
    total_kolesterol = 0
    for ingredient in row['food_ingredients_info']:
        total_air += ingredient.get('Air', 0)
        total_energi += ingredient.get('Energi', 0)
        total_protein += ingredient.get('Protein', 0)
        total_lemak += ingredient.get('Lemak', 0)
        total_abu += ingredient.get('Abu', 0)
        total_karbohidrat += ingredient.get('Karbohidrat', 0)
        total_serat_total += ingredient.get('Serat Total', 0)
        total_gula_total += ingredient.get('Gula Total', 0)
        total_kalsium += ingredient.get('Kalsium (Ca)', 0)
        total_besi += ingredient.get('Besi (Fe)', 0)
        total_magnesium += ingredient.get('Magnesium (Mg)', 0)
        total_fosfor += ingredient.get('Fosfor (P)', 0)
        total_kalium += ingredient.get('Kalium (K)', 0)
        total_natrium += ingredient.get('Natrium (Na)', 0)
        total_seng += ingredient.get('Seng (Zn)', 0)
        total_temBaga += ingredient.get('Tembaga (Cu)', 0)
        total_mangan += ingredient.get('Mangan (Mn)', 0)
        total_selenium += ingredient.get('Selenium (Se)', 0)
        total_vitamin_c += ingredient.get('Vitamin C', 0)
        total_tiamina += ingredient.get('Tiamina (B1)', 0)
        total_riboflavin += ingredient.get('Riboflavin (B2)', 0)
        total_niasin += ingredient.get('Niasin', 0)
        total_pantotenat += ingredient.get('Pantotenat (B5)', 0)
        total_vitamin_b6 += ingredient.get('Vitamin B6', 0)
        total_folat += ingredient.get('Folat Total (B9)', 0)
        total_kolina += ingredient.get('Kolina', 0)
        total_vitamin_b12 += ingredient.get('Vitamin B12', 0)
        total_vitamin_a_iu += ingredient.get('Vitamin A IU', 0)
        total_vitamin_a_rae += ingredient.get('Vitamin A RAE', 0)
        total_retinol += ingredient.get('Retinol', 0)
        total_a_karoten += ingredient.get('a-karoten', 0)
        total_b_karoten += ingredient.get('b-karoten', 0)
        total_b_kriptosantin += ingredient.get('b-kriptosantin', 0)
        total_likopen += ingredient.get('Likopen', 0)
        total_zeaksantin_lutein += ingredient.get('Zeaksantin + Lutein', 0)
        total_vitamin_e += ingredient.get('Vitamin E', 0)
        total_vitamin_d += ingredient.get('Vitamin D', 0)
        total_vitamin_d_iu += ingredient.get('Vitamin D IU', 0)
        total_vitamin_k += ingredient.get('Vitamin K', 0)
        total_lemak_jenuh += ingredient.get('Lemak Jenuh', 0)
        total_lemak_tunggal += ingredient.get('Lemak Tunggal', 0)
        total_lemak_ganda += ingredient.get('Lemak Ganda', 0)
        total_kolesterol += ingredient.get('Kolesterol', 0)

    # Add new columns of nutrition to recipes
    recipes.at[index, 'Air'] = '{:.3f}'.format(total_air).replace('.', ',')
    recipes.at[index, 'Energi'] = '{:.3f}'.format(total_energi).replace('.', ',')
    recipes.at[index, 'Protein'] = '{:.3f}'.format(total_protein).replace('.', ',')
    recipes.at[index, 'Lemak'] = '{:.3f}'.format(total_lemak).replace('.', ',')
    recipes.at[index, 'Abu'] = '{:.3f}'.format(total_abu).replace('.', ',')
    recipes.at[index, 'Karbohidrat'] = '{:.3f}'.format(total_karbohidrat).replace('.', ',')
    recipes.at[index, 'Serat Total'] = '{:.3f}'.format(total_serat_total).replace('.', ',')
    recipes.at[index, 'Gula Total'] = '{:.3f}'.format(total_gula_total).replace('.', ',')
    recipes.at[index, 'Kalsium (Ca)'] = '{:.3f}'.format(total_kalsium).replace('.', ',')
    recipes.at[index, 'Besi (Fe)'] = '{:.3f}'.format(total_besi).replace('.', ',')
    recipes.at[index, 'Magnesium (Mg)'] = '{:.3f}'.format(total_magnesium).replace('.', ',')
    recipes.at[index, 'Fosfor (P)'] = '{:.3f}'.format(total_fosfor).replace('.', ',')
    recipes.at[index, 'Kalium (K)'] = '{:.3f}'.format(total_kalium).replace('.', ',')
    recipes.at[index, 'Natrium (Na)'] = '{:.3f}'.format(total_natrium).replace('.', ',')
    recipes.at[index, 'Seng (Zn)'] = '{:.3f}'.format(total_seng).replace('.', ',')
    recipes.at[index, 'Tembaga (Cu)'] = '{:.3f}'.format(total_temBaga).replace('.', ',')
    recipes.at[index, 'Mangan (Mn)'] = '{:.3f}'.format(total_mangan).replace('.', ',')
    recipes.at[index, 'Selenium (Se)'] = '{:.3f}'.format(total_selenium).replace('.', ',')
    recipes.at[index, 'Vitamin C'] = '{:.3f}'.format(total_vitamin_c).replace('.', ',')
    recipes.at[index, 'Tiamina (B1)'] = '{:.3f}'.format(total_tiamina).replace('.', ',')
    recipes.at[index, 'Riboflavin (B2)'] = '{:.3f}'.format(total_riboflavin).replace('.', ',')
    recipes.at[index, 'Niasin'] = '{:.3f}'.format(total_niasin).replace('.', ',')
    recipes.at[index, 'Pantotenat (B5)'] = '{:.3f}'.format(total_pantotenat).replace('.', ',')
    recipes.at[index, 'Vitamin B6'] = '{:.3f}'.format(total_vitamin_b6).replace('.', ',')
    recipes.at[index, 'Folat Total (B9)'] = '{:.3f}'.format(total_folat).replace('.', ',')
    recipes.at[index, 'Kolina'] = '{:.3f}'.format(total_kolina).replace('.', ',')
    recipes.at[index, 'Vitamin B12'] = '{:.3f}'.format(total_vitamin_b12).replace('.', ',')
    recipes.at[index, 'Vitamin A IU'] = '{:.3f}'.format(total_vitamin_a_iu).replace('.', ',')
    recipes.at[index, 'Vitamin A RAE'] = '{:.3f}'.format(total_vitamin_a_rae).replace('.', ',')
    recipes.at[index, 'Retinol'] = '{:.3f}'.format(total_retinol).replace('.', ',')
    recipes.at[index, 'a-karoten'] = '{:.3f}'.format(total_a_karoten).replace('.', ',')
    recipes.at[index, 'b-karoten'] = '{:.3f}'.format(total_b_karoten).replace('.', ',')
    recipes.at[index, 'b-kriptosantin'] = '{:.3f}'.format(total_b_kriptosantin).replace('.', ',')
    recipes.at[index, 'Likopen'] = '{:.3f}'.format(total_likopen).replace('.', ',')
    recipes.at[index, 'Zeaksantin + Lutein'] = '{:.3f}'.format(total_zeaksantin_lutein).replace('.', ',')
    recipes.at[index, 'Vitamin E'] = '{:.3f}'.format(total_vitamin_e).replace('.', ',')
    recipes.at[index, 'Vitamin D'] = '{:.3f}'.format(total_vitamin_d).replace('.', ',')
    recipes.at[index, 'Vitamin D IU'] = '{:.3f}'.format(total_vitamin_d_iu).replace('.', ',')
    recipes.at[index, 'Vitamin K'] = '{:.3f}'.format(total_vitamin_k).replace('.', ',')
    recipes.at[index, 'Lemak Jenuh'] = '{:.3f}'.format(total_lemak_jenuh).replace('.', ',')
    recipes.at[index, 'Lemak Tunggal'] = '{:.3f}'.format(total_lemak_tunggal).replace('.', ',')
    recipes.at[index, 'Lemak Ganda'] = '{:.3f}'.format(total_lemak_ganda).replace('.', ',')
    recipes.at[index, 'Kolesterol'] = '{:.3f}'.format(total_kolesterol).replace('.', ',')
```
