```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import tensorflow as tf
```

```python
# Read dataset csv
lauk = pd.read_csv(r'Dataset\Augmented Data\augmented_lauk_with_nutrition.csv')
minum_snack = pd.read_csv(r'Dataset\Augmented Data\augmented_minum_snack_with_nutrition.csv')
minum_nutrisi = pd.read_csv(r'Dataset\Augmented Data\augmented_minum_with_nutrition.csv')
pokok_lauk = pd.read_csv(r'Dataset\Augmented Data\augmented_pokok_lauk_with_nutrition.csv')
sayur = pd.read_csv(r'Dataset\Augmented Data\augmented_sayur_with_nutrition.csv')
snack_lauk = pd.read_csv(r'Dataset\Augmented Data\augmented_snack_lauk_with_nutrition.csv')
snack_nutri = pd.read_csv(r'Dataset\Augmented Data\augmented_snack_with_nutrition.csv')
Dataset=pd.concat([lauk,minum_snack,minum_nutrisi,pokok_lauk,sayur,snack_lauk,snack_nutri],ignore_index=True)
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', 10)
Dataset.head(10)
```

```python
Dataset.to_csv('Food_Data_Merge.csv')
```

```python
lauk.columns
```

```python
Dataset.replace({',': ''}, regex=True, inplace=True)
Dataset = Dataset.apply(pd.to_numeric, errors='ignore')
Dataset.describe()
```

```python
features = Dataset.drop(['name'],axis=1)
labels = Dataset['name']
label_encoder = LabelEncoder()
labels_encoded = pd.Series(label_encoder.fit_transform(labels))
X_train, X_test, y_train, y_test = train_test_split(features, labels_encoded, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(scaler.transform(X_test),columns=X_test.columns, index=X_test.index)
X_train_scaled
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((X_train_scaled, y_train))
test_dataset = tf.data.Dataset.from_tensor_slices((X_test_scaled, y_test))
train_dataset
```
